```python

import re
x = True
# check pw length < 6   
def pwlen(passw):
    flag = 0
    if (len(passw)< 6): 
        flag = -1
    return flag
 # check length > 16     
def pwlen1(passw):
    flag = 0
    if (len(passw)> 16): 
        flag = -1
    return flag
    
# check [a-z] # check [A-Z]   
def checkaz(passw):
    flag = 0
    if not re.search("[a-z]",passw): 
        flag = -1
    if not re.search("[A-Z]",passw): 
        flag = -1
    
    return flag
    
# check [1-9]   
def check19(passw):
    flag = 0
    if not re.search("[1-9]",passw): 
        flag = -1
    return flag

# check [$#@]   
def checkSpecialChar(passw):
    flag = 0
    if not re.search("[$#@]",passw): 
        flag = -1
    return flag
    
# check [\s]   
def checkWhiteSpace(passw):
    flag =0
    if not re.search("[\s]",passw): 
        flag = -1
    return flag
    
while x:
    passw= "1Werweus@"
    #Minimum length 6 characters. 
    if pwlen(passw) == -1:
        flag = -1
        break
    #Maximum length 16 characters.
    elif pwlen1(passw) == -1:
        flag = -1
        break
    #At least 1 letter between [a-z] and 1 letter between [A-Z].
    elif checkaz(passw) == -1:
        flag = -1
        break
    #At least 1 number between [0-9].
    elif check19(passw) == -1:
        flag = -1
        break
     #At least 1 character from [$#@]. 
    elif checkSpecialChar(passw) == -1:
        flag = -1
        break
    # check [\s] white space
    elif checkWhiteSpace(passw) == 0:
        flag = -1
        break
    else: 
        flag = 0
        print("Valid Password") 
        x=False
        break
    

if x:
    print("Not a Valid Password")

```

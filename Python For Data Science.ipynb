```python
df1 = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
df1['HL_PCT'] = (df['High']- df['Close']) / df1['Close'] *100.0
```

```python
df1
```

```python
df1['PCT_change'] = (df['Close']- df['Open']) / df1['Open'] *100.0
```

```python
df1
```

```python
finDF = df1[['Close','HL_PCT', 'PCT_change', 'Volume']]
```

```python
finDF.head()
```

```python
# end of video 1
```

```python
import math
forecast_col = 'Close'
finDF.fillna(-9999, inplace=True)
forecast_out = int(math.ceil(0.01 * len(finDF)))
```

```python
forecast_out
```

```python
finDF['label'] = finDF[forecast_col].shift(-forecast_out)
```

```python
finDF.head()
```

```python
finDF.dropna(inplace = True)
finDF.tail()
```

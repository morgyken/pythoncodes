```python
import pandas as pd

def main():

    # Read the csv filename

    file_name = input("Enter the filename : ")     

    car_df = []

    # Read the csv file into a dataframe

    try:

        car_df = pd.read_csv(file_name)

    except:

        print("Can't open file")

    

    # Read the function index to call

    function = 1

    

    # Get the existing country values

    country_names = car_df['origin'].unique()

    while(function != 4):

        print("""

        (1) Get max mpg model

        (2) Get min mpg model

        (3) Get model year range

        (4) Exit\n""")

        function = int(input("Enter selection 1-4 : "))

        

        # Check if the given index is between 1 and 4

        if function in range(1,4):

            # Read the country name

            country_name = input("Enter the country of origin : ")

            print(country_name)

            # check for the country_name

            if country_name in country_names:

                if function == 1:

                    model_name = get_max_mpg_model(car_df, country_name)

                    print("Best model from " + country_name + " is : " + model_name)

                elif function == 2:

                    model_name = get_min_mpg_model(car_df, country_name)

                    print("Worst model from " + country_name + " is : " + model_name)

                elif function == 3:

                    # min_year, max_year = 0

                    min_year, max_year = get_model_year_range(car_df, country_name)

                    print("Model years range from 19" + str(min_year) + " to 19" + str(max_year))

                # print("Inside method calling")       

            else:

               function = 1

               print("Invalid country")

        else:

            if function == 4:

                break

            function = 1

            print("Invalid selection - must be a number from 1 to 4")
    return

def get_max_mpg_model(car_df, country_name):

  # get the cars for the given country name

  cars = car_df.loc[car_df['origin'] == country_name]

  # find the model name with max mpg

  car = cars.loc[cars['mpg'] == cars['mpg'].unique().max()]

  return car['name']

def get_min_mpg_model(car_df, country_name):

  # get the cars for the given country name

  cars = car_df.loc[car_df['origin'] == country_name]

  # find the model name with min mpg

  car = cars.loc[cars['mpg'] == cars['mpg'].unique().min()]

  return car['name']

def get_model_year_range(car_df, country_name):

  # get the cars for the given country name

  cars = car_df.loc[car_df['origin'] == country_name]

  # find the min and max year of the model

  year = cars['model_year'].unique()

  return year.min(), year.max()

if __name__ == "__main__":
    main()
```
